将大人的小说变成儿童读物

```python
import os
import re

def split_chapters(input_files, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    try:
        content = ""
        for input_file in input_files:
            with open(input_file, 'r', encoding='utf-8') as file:
                content += file.read()
    except FileNotFoundError:
        print(f"文件 {input_file} 不存在，请检查路径是否正确。")
        return
    
    # 使用正则表达式分割章节
    chapters = re.split(r'(?=^(?:番外)?第[一二三四五六七八九十百千万零0123456789]+章)', content, flags=re.MULTILINE)
    
    for i, chapter in enumerate(chapters):
        if chapter.strip():  # 跳过空章节
            # 尝试匹配章节标题
            match = re.search(r'^(?:番外)?第[一二三四五六七八九十百千万零0123456789]+章', chapter, flags=re.MULTILINE)
            if match:
                chapter_title = match.group().strip()  # 获取匹配到的标题
            else:
                chapter_title = f"未知章节_{i+1}"  # 如果未匹配到标题，使用默认标题
            # 构造输出文件路径
            output_file_name = f"{i+1}.txt"
            output_file_path = os.path.join(output_folder, output_file_name)
            with open(output_file_path, 'w', encoding='utf-8') as output_file:
                output_file.write(chapter.strip())
            print(f"已生成文件: {output_file_path}")

# 输入文件和输出文件夹路径
input_files = ['c:/Users/13802/code/notebook_on_everything/tongshu/雪中悍刀行(1-500章).txt','c:/Users/13802/code/notebook_on_everything/tongshu/雪中悍刀行(501-991章).txt']
output_folder = 'c:/Users/13802/code/notebook_on_everything/tongshu/chapters'

# 调用函数
split_chapters(input_files, output_folder)
```

```python
from http.client import REQUEST_TIMEOUT
from openai import OpenAI
import base64
import datetime

# 移除无效字符并优化注释
def send_chat_request(model, messages, temperature=0.7, max_tokens=-1, stream=False):
    # 创建OpenAI客户端
    client = OpenAI(
        api_key='your-api-key-here',  # 建议使用环境变量代替硬编码
        base_url='http://ollama.21cnai.com:30091/v1'
    )
    
    # 调用API生成响应
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature,
        max_tokens=max_tokens if max_tokens > 0 else None,
        stream=stream,
        timeout = 300
    )
    return response.choices[0].message.content
def split_and_combine(input_string, max_length=1000):
    # 按回车键拆分成一个列表
    lines = input_string.split('\n')
    
    combined_lines = []
    current_chunk = ""
    
    for line in lines:
        # 如果当前行加上新行后仍然小于等于最大长度，则添加到当前块
        if len(current_chunk) + len(line) + 1 <= max_length:
            if current_chunk:
                current_chunk += "\n"
            current_chunk += line
        else:
            # 否则，将当前块添加到结果列表，并开始新的块
            combined_lines.append(current_chunk)
            current_chunk = line
    
    # 添加最后一个块（如果存在）
    if current_chunk:
        combined_lines.append(current_chunk)

    
    return combined_lines
def modify_novel(input):
    text =input
    messages = [
        {
            "role": "user",
            "content": [
                {
                    "type": "text",
                    "text": f"你是一个小说修订专家，请你尽量保留以下小说原文内容，但将其中过度暴力的文字，改成合适的隐晦的方式来表达，使得整个上下文语意连贯，表达流畅，最后输出完整的修改合并后的整体内容：\n{text}"
                },
            ]
        }
    ]
    response_content = send_chat_request(model="mistral-small-3.1-instruct-2503", messages=messages)
    return response_content

def rewrite_chapter(file_path):
    try:
        # 读取文件内容
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
        result_array = split_and_combine(content)
        rewritten_content = ""
        for result in result_array:
            # 使用modify_novel方法进行改写
            rewritten_content += modify_novel(result)
            print(rewritten_content)
        
        # 打印或返回改写后的内容
        return rewritten_content
    
    except FileNotFoundError:
        print(f"文件 {file_path} 不存在，请检查路径是否正确。")
    except Exception as e:
        print(f"发生错误: {e}")

# 调用函数
file_path = 'C:/Users/13802/code/notebook_on_everything/tongshu/chapters/871.txt'
print(rewrite_chapter(file_path))
```
